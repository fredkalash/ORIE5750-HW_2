Q2a) Consider a patient with fever and no headache.

There are 0 such patients Pneumonia class.

There are 6 such patients in the Flu class.

There are 3 such patients in the "healthy" class. 

So the Prob(Pneumonia | fever and no headache)  =  0

Prob(Flu | fever and no headache)  =  6 / (0+6+3) = 6/9 = 2/3

Prob(Healthy | fever and no headache) = 3 / (0+6+3) = 3/9 = 1/3

So the Bayes Optimal Classifier will pick Flu with probability 2/3.
